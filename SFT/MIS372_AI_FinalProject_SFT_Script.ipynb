```python
import torch
from unsloth import FastLanguageModel
from datasets import load_dataset
from trl import SFTTrainer, SFTConfig
from transformers import DataCollatorForSeq2Seq, TextStreamer
import os

# --- Configuration ---
MODEL_NAME = "unsloth/Llama-3.2-1B-Instruct"
DATASET_PATH = "sft_dataset.jsonl"
OUTPUT_DIR_TRAINING = "outputs/scriptwriter_model"
OUTPUT_DIR_MODEL = "models/fortnite_scriptwriter"
MAX_SEQ_LENGTH = 2048
LORA_RANK = 16
LORA_ALPHA = 32
TRAINING_STEPS = 200
LEARNING_RATE = 1e-4
# ---------------------
```

```python
def train_sft_model():
    """
    This function handles the supervised fine-tuning (SFT) of a Llama-3.2 model
    to convert Fortnite patch note summaries into YouTube video scripts.
    """
    # 1. Load the Base Model
    load_in_4bit = True
    dtype = None

    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name=MODEL_NAME,
        max_seq_length=MAX_SEQ_LENGTH,
        dtype=dtype,
        load_in_4bit=load_in_4bit,
    )
    print("Model and tokenizer loaded successfully.")

    # 2. PEFT: Adding LoRA Adapters
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "up_proj", "down_proj"]
    lora_dropout = 0
    lora_model = FastLanguageModel.get_peft_model(
        model,
        r=LORA_RANK,
        target_modules=target_modules,
        lora_alpha=LORA_ALPHA,
        lora_dropout=lora_dropout,
        bias="none",
        use_gradient_checkpointing="unsloth",
        random_state=3407,
        use_rslora=False,
    )
    trainable_params = sum(p.numel() for p in lora_model.parameters() if p.requires_grad)
    print(f"Trainable parameters: {trainable_params:,}")

    # 3. Data Pipeline
    if not os.path.exists(DATASET_PATH):
        raise FileNotFoundError(f"Dataset file not found at {DATASET_PATH}. Please ensure it's been uploaded.")
    dataset = load_dataset("json", data_files=DATASET_PATH, split="train")

    from unsloth.chat_templates import get_chat_template
    tokenizer = get_chat_template(
        tokenizer,
        chat_template="llama-3.1",
    )

    def format_dataset(example):
        return {
            "conversations": [
                {"role": "user", "content": f"Please convert the following Fortnite patch notes summary into a humorous and engaging YouTube video script. Make sure to explicitly mention all the areas from summary that had 'No information provided.', e.g. Map Changes, Gameplay, Cosmetics & Collabs, etc.:\n\n{example['input']}"},
                {"role": "assistant", "content": example['output']}
            ]
        }
    formatted_dataset = dataset.map(format_dataset)

    def formatting_prompts_func(examples):
        convos = examples["conversations"]
        texts = [tokenizer.apply_chat_template(convo, tokenize=False, add_generation_prompt=False) for convo in convos]
        return {"text": texts}
    processed_dataset = formatted_dataset.map(formatting_prompts_func, batched=True)
    print("Dataset loaded and formatted.")

    # 4. Training
    batch_size = 2
    gradient_accumulation_steps = 4
    use_fp16 = torch.cuda.is_available() and not torch.cuda.is_bf16_supported()
    use_bf16 = torch.cuda.is_available() and torch.cuda.is_bf16_supported()

    trainer = SFTTrainer(
        model=lora_model,
        tokenizer=tokenizer,
        train_dataset=processed_dataset,
        dataset_text_field="text",
        max_seq_length=MAX_SEQ_LENGTH,
        data_collator=DataCollatorForSeq2Seq(tokenizer=tokenizer),
        packing=False,
        args=SFTConfig(
            per_device_train_batch_size=batch_size,
            gradient_accumulation_steps=gradient_accumulation_steps,
            warmup_steps=10,
            max_steps=TRAINING_STEPS,
            learning_rate=LEARNING_RATE,
            fp16=use_fp16,
            bf16=use_bf16,
            logging_steps=10,
            optim="adamw_8bit",
            weight_decay=0.01,
            lr_scheduler_type="linear",
            seed=3407,
            output_dir=OUTPUT_DIR_TRAINING,
            report_to="none",
        ),
    )

    from unsloth.chat_templates import train_on_responses_only
    instruction_part = "<|start_header_id|>user<|end_header_id|>\n\n"
    response_part = "<|start_header_id|>assistant<|end_header_id|>\n\n"
    trainer = train_on_responses_only(
        trainer,
        instruction_part=instruction_part,
        response_part=response_part,
    )

    print("Trainer configured. Starting training...")
    trainer.train()
    print("Training complete.")

    # 5. Save the Model
    trainer.save_model(OUTPUT_DIR_MODEL)
    tokenizer.save_pretrained(OUTPUT_DIR_MODEL)
    print(f"Model saved to {OUTPUT_DIR_MODEL}")

# Run the training process
train_sft_model()
```

```python
import io
import sys

def generate_script(patch_summary, model_path=OUTPUT_DIR_MODEL):
    """
    Loads the fine-tuned model and generates a YouTube script from a patch summary.
    """
    if not os.path.exists(model_path):
        print(f"Model not found at {model_path}. Please train the model first.")
        return

    print("\n--- Generating Script ---")

    model, tokenizer = FastLanguageModel.from_pretrained(
        model_name=model_path,
        max_seq_length=MAX_SEQ_LENGTH,
        dtype=None,
        load_in_4bit=True,
    )
    FastLanguageModel.for_inference(model)

    device = next(model.parameters()).device

    messages = [
        {"role": "user", "content": f"Please convert the following Fortnite patch notes summary into a humorous and engaging YouTube video script. When you see areas in the patch notes summary that contain 'No information provided.', assume that no updates/changes were made to those specific areas and make sure to name drop these areas (e.g., Map Changes and Gameplay Changes, etc.):\n\n{patch_summary}"}
    ]

    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to(device)

    # Redirect stdout to capture the streamer's output
    original_stdout = sys.stdout
    captured_output = io.StringIO()
    sys.stdout = captured_output

    text_streamer = TextStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)

    print("Generated Script:")
    _ = model.generate(
        input_ids=inputs,
        streamer=text_streamer,
        max_new_tokens=512,
        use_cache=True,
        temperature=0.7,
        top_p=0.95,
    )

    # Restore stdout and get the captured text
    sys.stdout = original_stdout
    generated_text = captured_output.getvalue()

    # Print the captured text to the actual console
    print(generated_text)

    # Save the generated script to a file
    output_filename = os.path.join(OUTPUT_DIR_MODEL, "generated_script.txt")
    os.makedirs(OUTPUT_DIR_MODEL, exist_ok=True)
    with open(output_filename, "w") as f:
        f.write(generated_text)
    print(f"\nScript saved to {output_filename}")

# Example to test the generation
example_summary = """
**Fortnite Patch Summary**

### ğŸ“ Overview
- Multiple updates including a hotfix and several patches.
- Homer Simpson-related events involving Zero Point energy and a giant Homer cloud.
- Donut Storm mechanics introduced.
- Springfield Island affected with Zero Point energy effects.
- Storm zones temporarily changed to shape of the number 7.
- KBBL Radio Towers disabled and re-enabled.
- Current loot pool includes various weapons and utility items, with some vaulted and added items.

### ğŸ—ºï¸ Map Changes
- No information provided.

### ğŸ”« Weapons / Loot Pool
- Striker AR
- Suppressed Assault Rifle
- Infantry Rifle
- Tactical Shotgun
- Thunder Shotgun
- Ranger Shotgun
- Five Gun
- Compact SMG
- Machine SMG
- Deodorant Applicator
- Mr. Blasty
- Suppressed Pistol
- Cobalt Slingshot
- Cartoon Chainsaw
- Bandage
- Med Kit
- Shield Potion
- Small Shield Potion
- Chug Splash
- Chug Jug
- Krusty Burger
- Super Squishee
- Blinky Fish
- Flopper
- Small Fry
- Fishing Rod
- Shockwave Grenade
- Pulse Scanner
- Port-A-Bunker
- Port-A-Cover
- Big Bush Bomb
- Small Fries
- Shield Fish
- Cabbage
- Corn
- Banana
- Coconut
- Apple
- Shield Mushroom
- Tomato

### ğŸ® Gameplay Changes
- No information provided.

### ğŸ‘• Cosmetics & Collabs
- The Simpsons cast, including Homer Simpson, Marge Simpson, Lisa Simpson, Bart Simpson, and Maggie Simpson.

### ğŸ“… Quests & Events
- Homer Simpson used Zero Point on a donut-shaped cloud to make it rain donuts.
- Donut Storm may form after a Storm phase.
- Homer Simpson cloud evolves from a cloud to a giant Homer Simpson figure.
- Springfield Nuclear Power Plant erupts with Zero Point energy.
- A countdown activated in Springfield Town Square.
- Storm zones changed to shape of the number 7, then reverted.
- KBBL Radio Towers disabled and re-enabled.

### âš¡ TLDR
Multiple updates introduced Homer Simpson-themed events, donut storms, and Zero Point effects on Springfield Island. The loot pool includes various weapons and utility items, with some vaulted and added items. Map changes and gameplay adjustments are not specified.
"""
generate_script(example_summary)
```

```python
import replicate
import os
from google.colab import userdata
import requests # Import the requests library

# Ensure the Replicate API token is available
REPLICATE_API_TOKEN = userdata.get('REPLICATE_API_TOKEN')
if not REPLICATE_API_TOKEN:
    raise ValueError("REPLICATE_API_TOKEN not found in Colab secrets. Please add it.")
os.environ["REPLICATE_API_TOKEN"] = REPLICATE_API_TOKEN

# Path to the generated script
script_path = os.path.join(OUTPUT_DIR_MODEL, "generated_script.txt")

if not os.path.exists(script_path):
    print(f"Generated script not found at {script_path}. Please generate the script first.")
else:
    with open(script_path, "r") as f:
        generated_script_content = f.read()

    print("\n--- Generating Videos using Sora-2 ---")
    print("This process might take a while, depending on the number of lines and Replicate's queue.")

    # Parse the script content
    script_lines = []
    if "Generated Script:" in generated_script_content:
        # Split the content by lines and find the start of the actual script
        all_lines = generated_script_content.split('\n')
        header_found = False
        for line in all_lines:
            if "Generated Script:" in line:
                header_found = True
                continue
            if header_found:
                stripped_line = line.strip()
                if stripped_line:
                    script_lines.append(stripped_line)
    else:
        # If header not found, assume the whole content is the script, split by lines
        for line in generated_script_content.split('\n'):
            stripped_line = line.strip()
            if stripped_line:
                script_lines.append(stripped_line)

    if not script_lines:
        print("No valid script lines found to generate videos.")
    else:
        for i, line_content in enumerate(script_lines):
            # Ensure the line_content is properly formatted with escaped quotes for the f-string
            prompt_for_sora = f"A Fortnite Youtuber looking at the camera and saying: \"{line_content}\""
            print(f"\nAttempting to generate video for line {i+1}: {prompt_for_sora[:70]}...")

            try:
                # The replicate.run function directly returns the output (e.g., a URL string)
                video_url = replicate.run(
                    "openai/sora-2",
                    input={
                        "prompt": prompt_for_sora,
                        "seconds": 8, # Adjust video length as needed
                        "aspect_ratio": "landscape"
                    }
                )

                if video_url:
                    print(f"Video {i+1} generated successfully! You can view it here: {video_url}")

                    # Download the video content using requests
                    video_response = requests.get(video_url)
                    video_response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

                    video_filename = os.path.join(OUTPUT_DIR_MODEL, f"generated_video_part_{i+1}.mp4")
                    with open(video_filename, "wb") as f:
                        f.write(video_response.content)
                    print(f"Video {i+1} saved to {video_filename}")
                else:
                    print(f"Failed to generate video or get URL for line {i+1}.")

            except Exception as e:
                print(f"An error occurred during video generation for line {i+1}: {e}")
```

```python
from moviepy.editor import VideoFileClip, concatenate_videoclips
import os
import re

def stitch_videos(output_dir=OUTPUT_DIR_MODEL, output_filename="final_video.mp4"):
    print("\n--- Stitching Videos ---")

    # Find all generated video files
    video_files = [f for f in os.listdir(output_dir) if f.startswith("generated_video_part_") and f.endswith(".mp4")]

    if not video_files:
        print("No video parts found to stitch.")
        return

    # Sort files by the numeric part in the filename to ensure correct order
    # Filename format expected: generated_video_part_{i}.mp4
    video_files.sort(key=lambda f: int(re.search(r'part_(\d+)', f).group(1)))

    print(f"Found {len(video_files)} video parts to stitch.")

    clips = []
    for video_file in video_files:
        file_path = os.path.join(output_dir, video_file)
        try:
            clip = VideoFileClip(file_path)
            clips.append(clip)
        except Exception as e:
            print(f"Error loading clip {file_path}: {e}")

    if clips:
        try:
            # Concatenate all clips
            final_clip = concatenate_videoclips(clips)
            output_path = os.path.join(output_dir, output_filename)

            # Write the result to a file
            final_clip.write_videofile(output_path, codec="libx264", audio_codec="aac")
            print(f"\nSuccessfully stitched video saved to: {output_path}")
        except Exception as e:
            print(f"Error concatenating clips: {e}")
        finally:
             # Close clips to release resources
            for clip in clips:
                clip.close()
            if 'final_clip' in locals():
                final_clip.close()
    else:
        print("No valid clips loaded.")

# Run the stitching function
stitch_videos()
```
